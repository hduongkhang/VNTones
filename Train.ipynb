```
# -*- coding: utf-8 -*-
!pip install unidecode
try:
    import re2 as re
except ImportError:
    import re
import os
import unidecode
from tqdm import tqdm
from math import log
import numpy as np
import itertools
from nltk import ngrams
import string
import pickle

MAXLEN = 30
NGRAM = 5
# BATCH_SIZE = 1024


# RIGHT2LEFT = True
```

```
#Hàm xóa dấu câu đầu vào
def remove_accent(text):
    return unidecode.unidecode(text)

#Split words tao NGRAM word
def gen_ngrams(words, n=5):
    return ngrams(words.split(), n)

def extract_phrases(text):
    return re.findall(r'\w[\w ]+', text)


alphabet_match = '^[ _abcdefghijklmnopqrstuvwxyz0123456789áàảãạâấầẩẫậăắằẳẵặóòỏõọôốồổỗộơớờởỡợéèẻẽẹêếềểễệúùủũụưứừửữựíìỉĩịýỳỷỹỵđ!\"\',\-\.:;?_\(\)]+$'
```

```
list_ngrams = []

directory = os.fsencode('/content/drive/My Drive/VNTones/train2/')   
for file in os.listdir(directory):
  filename = os.fsdecode(file)
  if filename == "bj" or filename == "aj" or filename == "split_file0007" or filename == "split_file0012" or filename == "split_file0001":
    list_ngrams = list(set(list_ngrams))
  newPathfile = '/content/drive/My Drive/VNTones/train2/' + filename
  print(newPathfile)
  #newPathfile = '/content/drive/My Drive/Thesis/test.txt'
  # -*- coding: utf-8 -*-
  with open(newPathfile,'r',encoding = 'utf-8',errors='replace') as f_r:
    lines = f_r.read().split("\n")
    # if RIGHT2LEFT:
    #   lines = [x[::-1] for x in lines]

  #Tách câu thành các pharse bằng cách tách theo dấu câu

  phrases = itertools.chain.from_iterable(extract_phrases(text) for text in lines)
  phrases = [p.strip() for p in phrases if len(p.split()) > 1]
  del lines
  for p in tqdm(phrases):
    if not re.match(alphabet_match, p.lower()):
      continue
    for ngr in gen_ngrams(p, NGRAM):
      if len(" ".join(ngr)) < 30:
        list_ngrams.append(" ".join(ngr))
  del phrases
list_ngrams = list(set(list_ngrams))

# open a file, where you ant to store the data
file = open('/content/drive/My\ Drive/VNTones/huge_list_ngrams.pkl', 'wb')

# dump information to that file
pickle.dump(list_ngrams, file)

# close the file
file.close()
```

```


accented_chars_vietnamese = [
    'á', 'à', 'ả', 'ã', 'ạ', 'â', 'ấ', 'ầ', 'ẩ', 'ẫ', 'ậ', 'ă', 'ắ', 'ằ', 'ẳ', 'ẵ', 'ặ',
    'ó', 'ò', 'ỏ', 'õ', 'ọ', 'ô', 'ố', 'ồ', 'ổ', 'ỗ', 'ộ', 'ơ', 'ớ', 'ờ', 'ở', 'ỡ', 'ợ',
    'é', 'è', 'ẻ', 'ẽ', 'ẹ', 'ê', 'ế', 'ề', 'ể', 'ễ', 'ệ',
    'ú', 'ù', 'ủ', 'ũ', 'ụ', 'ư', 'ứ', 'ừ', 'ử', 'ữ', 'ự',
    'í', 'ì', 'ỉ', 'ĩ', 'ị',
    'ý', 'ỳ', 'ỷ', 'ỹ', 'ỵ',
    'đ',
]
accented_chars_vietnamese.extend([c.upper() for c in accented_chars_vietnamese])
alphabet = list(('\x00 _' + string.ascii_letters + string.digits + ''.join(accented_chars_vietnamese)))


#Model
from keras.models import Sequential
from keras.layers import CuDNNGRU, CuDNNLSTM, GRU,Activation, TimeDistributed, Dense, RepeatVector, recurrent, LSTM, Bidirectional
from keras.callbacks import Callback, EarlyStopping, ModelCheckpoint
from keras.optimizers import Adam
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences

HIDDEN_SIZE = 256

model = Sequential()
model.add(LSTM(HIDDEN_SIZE, input_shape=(MAXLEN, len(alphabet)), return_sequences=True))
model.add(Bidirectional(LSTM(HIDDEN_SIZE, return_sequences=True, dropout=0.25, recurrent_dropout=0.1)))
model.add(Dense(len(alphabet),activation='relu'))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer=Adam(lr=0.001),
              metrics=['accuracy'])
'''
model = Sequential()
model.add(GRU(HIDDEN_SIZE, input_shape=(MAXLEN, len(alphabet)), return_sequences=True))
model.add(Bidirectional(GRU(HIDDEN_SIZE, return_sequences=True, dropout=0.25, recurrent_dropout=0.1)))
model.add(Dense(128,activation='relu'))
model.add(Dense(len(alphabet), activation='softmax'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
'''

model.summary()


#encode and decoder
def encode(text, maxlen=MAXLEN):
        text = "\x00" + text
        # print(text)
        x = np.zeros((maxlen, len(alphabet)))
        for i, c in enumerate(text[:maxlen]):
            x[i, alphabet.index(c)] = 1
        if i < maxlen - 1:
          for j in range(i+1, maxlen):
            x[j, 0] = 1
        return x
#dùng one-hot vector định nghĩa, ra được 1 matrix one-vector mỗi row là 1 char (len = len of alphabet) và 1 matrix là 1ngram maxlen=30.
#tôi => t [ 0 0 0 1 0 ... 0]
#       ô [ 0 1 0 0 0 ... 0]
#       i [ 0 0 1 0 0 ... 0]




def greedy_decoder(data):
	# index for largest probability each row
	return [argmax(s) for s in data]   
      
def decode(x, calc_argmax=True):
    if calc_argmax:
        x = x.argmax(axis=-1)
    return ''.join(alphabet[i] for i in x)


#phân chia tập dữ liệu validate
from sklearn.model_selection import train_test_split
train_data, valid_data = train_test_split(list_ngrams_pkl, test_size=0.2, random_state=2020)


def generate_data(data, batch_size=128):
    cur_index = 0
    while True:
        
        x, y = [], []
        for i in range(batch_size):  
            y.append(encode(data[cur_index]))
            x.append(encode(remove_accent(data[cur_index])))
            cur_index += 1
            
            if cur_index > len(data)-1:
                cur_index = 0
        
        yield np.array(x), np.array(y)

#check lại epoch cũ
path_checkpoint = '/content/drive/My Drive/Thesis/model'
if (os.path.exists(path_checkpoint)):
    listEpochs = [x for x in os.listdir(path_checkpoint) if x[:6] == 'model_' and x[-3:] == '.h5']

    if (len(listEpochs) > 0):
        lastEpoch = max([int(x[-6:-3]) for x in listEpochs])
        lastEpochFile = path_checkpoint + '/' + 'model_%03d.h5' % (lastEpoch)
        # load weights
        model.load_weights(lastEpochFile)
        print("CONTINUE TRAINING FROM EPOCH %03d......" % (lastEpoch))
    else:
        lastEpoch = 0
#training work
train_generator = generate_data(train_data, batch_size=BATCH_SIZE)
validation_generator = generate_data(valid_data, batch_size=BATCH_SIZE)

checkpointer = ModelCheckpoint(filepath=os.path.join(path_checkpoint,'model_{epoch:03d}.h5'), verbose=1, period = 1)
early = EarlyStopping(patience=2, verbose=1)

model.fit_generator(train_generator, steps_per_epoch=len(train_data)//BATCH_SIZE, epochs=50,
                    validation_data=validation_generator, validation_steps=len(valid_data)//BATCH_SIZE,
                    callbacks=[checkpointer, early], initial_epoch = lastEpoch)
```
